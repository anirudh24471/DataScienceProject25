```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report

# =========================================================
# CONFIG
# =========================================================
X_path = "X"
aug_path = "augmented_64"
y_path = "Y.npy"
test_path = "test_data"
y1_test_path = "Y1.npy"
model_save_path = "trained_3dcnn.keras"

feature_names = [
    "Oring", "circular_end_pocket", "v_circular_end_blind_slot",
    "circular_pocket", "non_rectangular_pocket", "rectangular_pocket", "triangular_pocket"
]
num_classes = len(feature_names)

# =========================================================
# LABEL ENCODER
# =========================================================
y = np.load(y_path)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
np.save("label_classes.npy", le.classes_)

# =========================================================
# LOAD TRAINING DATA
# =========================================================
X_list = []
for i, name in enumerate(feature_names):
    f = os.path.join(X_path, f"{i}_{name}_concatenated.npy")
    X_list.append(np.load(f))
X = np.concatenate(X_list, axis=0)[..., np.newaxis]
X = X[:len(y_encoded)]

augmented_list, aug_labels = [], []
for i, name in enumerate(feature_names):
    f = os.path.join(aug_path, f"{i}_{name}_voxel_grid_augmented.npy")
    if os.path.exists(f):
        d = np.load(f)
        augmented_list.append(d)
        aug_labels.extend([i] * len(d))

augmented_data = np.concatenate(augmented_list, axis=0)[..., np.newaxis]
aug_labels = np.array(aug_labels)

X_total = np.concatenate([X, augmented_data], axis=0)
y_total = np.concatenate([y_encoded, aug_labels], axis=0)

# =========================================================
# BALANCE CLASSES
# =========================================================
def balance_classes(X, y, num_per_class=3000, seed=42):
    np.random.seed(seed)
    X_bal, y_bal = [], []
    for cls in np.unique(y):
        indices = np.where(y == cls)[0]
        np.random.shuffle(indices)
        if len(indices) >= num_per_class:
            selected = indices[:num_per_class]
        else:
            selected = np.random.choice(indices, num_per_class, replace=True)
        X_bal.append(X[selected])
        y_bal.append(y[selected])
    X_final = np.concatenate(X_bal, axis=0)
    y_final = np.concatenate(y_bal, axis=0)
    p = np.random.permutation(len(y_final))
    return X_final[p], y_final[p]

X_total, y_total = balance_classes(X_total, y_total, num_per_class=3000)

# Train/Val split
X_train, X_val, y_train, y_val = train_test_split(
    X_total, y_total, test_size=0.2, stratify=y_total, random_state=42
)

# =========================================================
# MODEL DEFINITION
# =========================================================
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
        print(e)

# Use float32 to avoid MaxPool3D float16 error on CPU
tf.keras.mixed_precision.set_global_policy('float32')

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(64, 64, 64, 1)),
    tf.keras.layers.Conv3D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling3D(2),
    tf.keras.layers.Conv3D(64, 3, activation='relu'),
    tf.keras.layers.MaxPooling3D(2),
    tf.keras.layers.Conv3D(128, 3, activation='relu'),
    tf.keras.layers.GlobalAveragePooling3D(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# =========================================================
# TRAINING
# =========================================================
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint("best_model.keras", monitor='val_accuracy', save_best_only=True)
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=10,
    batch_size=4,
    callbacks=callbacks,
)

model.save(model_save_path)
print(f"\n Model saved to {model_save_path}")

# =========================================================
# PLOT TRAINING
# =========================================================
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label="Train")
plt.plot(history.history['val_accuracy'], label="Val")
plt.title("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label="Train")
plt.plot(history.history['val_loss'], label="Val")
plt.title("Loss")
plt.legend()
plt.tight_layout()
plt.show()

# =========================================================
# TESTING
# =========================================================
test_X_files = sorted([f for f in os.listdir(test_path) if f.endswith(".npy")])
X_test_list = []
for f in test_X_files:
    arr = np.load(os.path.join(test_path, f))
    if arr.ndim == 3:
        arr = arr[np.newaxis, ...]
    X_test_list.append(arr)

X_test_final = np.concatenate(X_test_list, axis=0)[..., np.newaxis].astype(np.float32)
y_test_final = np.load(y1_test_path)

# Predict
y_pred_probs = model.predict(X_test_final, batch_size=8)
y_pred = np.argmax(y_pred_probs, axis=1)

# Labels
cm_labels = feature_names
true_names = [cm_labels[i] for i in y_test_final]
pred_names = [cm_labels[i] for i in y_pred]

# =========================================================
# PRINT PREDICTIONS
# =========================================================
print("\n\U0001F9FE Test Sample Predictions:")
print(f"{'Index':<6} {'True Label':<30} {'Predicted Label'}")
print("-" * 60)
for i, (t, p) in enumerate(zip(true_names, pred_names)):
    print(f"{i:<6} {t:<30} {p}")
```

```python
import os, math, random
import numpy as np
import tensorflow as tf

# ----------- QUIETER TF LOGS (optional) -----------
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

# ----------- GPU SETUP -----------
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for g in gpus:
            tf.config.experimental.set_memory_growth(g, True)
        print(f"Using GPU(s): {[tf.config.experimental.get_device_details(g)['device_name'] for g in gpus]}")
    except Exception as e:
        print("GPU memory growth not set:", e)
else:
    print("No GPU detected  will train on CPU.")

# ----------- CONFIG -----------
X_DIR      = "X"
AUG_DIR    = "augmented_64"
Y_PATH     = "Y.npy"
FEATURES   = [
    "Oring", "circular_end_pocket", "v_circular_end_blind_slot",
    "circular_pocket", "non_rectangular_pocket", "rectangular_pocket", "triangular_pocket"
]
VOXEL_SIZE = 64           # input resolution on disk
CHANNELS   = 1
INPUT_SHAPE = (VOXEL_SIZE, VOXEL_SIZE, VOXEL_SIZE, CHANNELS)

SEED        = 42
BATCH_SIZE  = 2           # 64^3 is heavy; start with 24
EPOCHS      = 12
VAL_SPLIT   = 0.2
SHUFFLE_BUF = 8192
DOWNSAMPLE_TO_32 = False  # set True to reduce to 32^3 inside pipeline

rng = np.random.default_rng(SEED)
tf.keras.utils.set_random_seed(SEED)

# ----------- LABELS -----------
y = np.load(Y_PATH)                          # shape (N,)
# we will use integer labels already in Y.npy (0..6)
assert y.ndim == 1, "Y.npy must be 1D of integer class ids."

# ----------- MEMORY-MAPPED DATA SOURCES -----------
# Each .npy contains (N_i, 64,64,64) uint8 or {0,1}
# We open them with mmap_mode='r' so they don't load fully into RAM.
class NpyBank:
    def __init__(self):
        self.mem = {}  # key -> memmap array
        self.lengths = {}  # key -> len
        self.labels = {}   # key -> class_id

    def add(self, key, path, label):
        arr = np.load(path, mmap_mode='r')  # memmap
        assert arr.ndim == 4 and arr.shape[1:]==(VOXEL_SIZE,VOXEL_SIZE,VOXEL_SIZE), f"Bad shape in {path}: {arr.shape}"
        self.mem[key] = arr
        self.lengths[key] = arr.shape[0]
        self.labels[key] = label

    def __len__(self):
        return sum(self.lengths.values())

    def get(self, key, idx):
        # returns (H,W,D) uint8 or 0/1
        return self.mem[key][idx]

# Load perfect data files
bank = NpyBank()
for i, name in enumerate(FEATURES):
    f = os.path.join(X_DIR, f"{i}_{name}_concatenated.npy")
    if not os.path.exists(f):
        raise FileNotFoundError(f)
    bank.add(f"X/{i}", f, i)

# Load augmented data files if present
for i, name in enumerate(FEATURES):
    f = os.path.join(AUG_DIR, f"{i}_{name}_voxel_grid_augmented.npy")
    if os.path.exists(f):
        bank.add(f"A/{i}", f, i)

print(f"Total samples across all files (memmap): {len(bank):,}")

# ----------- BUILD GLOBAL INDEX -----------
# We create a flat list of (key, local_index, label) for all samples.
index = []
for key, length in bank.lengths.items():
    label = bank.labels[key]
    index.extend([(key, j, label) for j in range(length)])

# Optionally align to the length of Y if you insist on matching original Y count.
# In practice, were training supervised on (bank, labels) we created, so this is optional.

# Shuffle once for split
rng.shuffle(index)

# Split
val_count = int(len(index) * VAL_SPLIT)
val_index = index[:val_count]
train_index = index[val_count:]

print(f"Train: {len(train_index):,} | Val: {len(val_index):,}")

# ----------- UTILS -----------
def to_minus1_plus1(x):
    x = x.astype(np.float32)
    if x.max() > 1.0:
        x *= (1.0/255.0)
    return x * 2.0 - 1.0

def load_from_index_py(key, idx):
    """Python side loader for tf.numpy_function."""
    vol = bank.get(key.decode('utf-8'), int(idx))  # (64,64,64) uint8/0-1
    vol = to_minus1_plus1(vol)                     # [-1,1]
    if DOWNSAMPLE_TO_32:
        # simple 2x2x2 average downsample to 32^3
        v = vol.reshape(32,2,32,2,32,2).mean(axis=(1,3,5))
        vol = v.astype(np.float32)
    return np.expand_dims(vol, axis=-1)            # (H,W,D,1)

def tf_loader(key, idx, label):
    vol = tf.numpy_function(load_from_index_py, [key, idx], tf.float32)
    target_shape = (32,32,32,1) if DOWNSAMPLE_TO_32 else (64,64,64,1)
    vol.set_shape(target_shape)
    return vol, label

# Convert index to tensors
def make_dataset(pairs, batch, shuffle, repeat=False):
    keys  = np.array([p[0].encode('utf-8') for p in pairs])
    idxs  = np.array([p[1] for p in pairs], dtype=np.int64)
    labs  = np.array([p[2] for p in pairs], dtype=np.int32)

    ds = tf.data.Dataset.from_tensor_slices((keys, idxs, labs))
    if shuffle:
        ds = ds.shuffle(min(len(pairs), SHUFFLE_BUF), seed=SEED, reshuffle_each_iteration=True)
    ds = ds.map(tf_loader, num_parallel_calls=tf.data.AUTOTUNE)
    if repeat:
        ds = ds.repeat()
    ds = ds.batch(batch, drop_remainder=False).prefetch(tf.data.AUTOTUNE)
    return ds

train_ds = make_dataset(train_index, BATCH_SIZE, shuffle=True, repeat=False)
val_ds   = make_dataset(val_index,   BATCH_SIZE, shuffle=False, repeat=False)

# ----------- MIXED PRECISION -----------
from tensorflow.keras import mixed_precision
if gpus:
    mixed_precision.set_global_policy('mixed_float16')
else:
    mixed_precision.set_global_policy('float32')

# ----------- FEATURNET-LIKE MODEL -----------
from tensorflow.keras import layers, models

def build_featurenet(input_shape, num_classes=7):
    inputs = layers.Input(shape=input_shape)

    x = layers.Conv3D(32, 7, strides=2, padding='same', use_bias=False)(inputs)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)

    x = layers.Conv3D(32, 5, strides=1, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)

    x = layers.Conv3D(64, 4, strides=1, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)

    x = layers.Conv3D(64, 3, strides=1, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x); x = layers.ReLU()(x)

    x = layers.MaxPool3D(pool_size=2, strides=2)(x)

    # Optional: a small extra block helps when downsampling to 32
    if DOWNSAMPLE_TO_32:
        x = layers.Conv3D(96, 3, padding='same', use_bias=False)(x)
        x = layers.BatchNormalization()(x); x = layers.ReLU()(x)

    x = layers.Flatten()(x)
    x = layers.Dense(128)(x); x = layers.ReLU()(x); x = layers.Dropout(0.35)(x)

    # Keep logits in float32 when using mixed precision
    outputs = layers.Dense(num_classes, activation='softmax', dtype='float32')(x)
    return models.Model(inputs, outputs, name="FeatureNet3D")

effective_input = (32,32,32,1) if DOWNSAMPLE_TO_32 else (64,64,64,1)
model = build_featurenet(effective_input, num_classes=7)
model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

# ----------- CALLBACKS -----------
ckpt = tf.keras.callbacks.ModelCheckpoint(
    "featurenet_best.keras", monitor="val_accuracy", save_best_only=True, verbose=1
)
early = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy", patience=5, restore_best_weights=True, verbose=1
)
plateau = tf.keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss", factor=0.5, patience=3, verbose=1
)

# ----------- TRAIN -----------
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=[ckpt, early, plateau],
    verbose=1
)

# ----------- PLOTS -----------
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='val')
plt.title('Accuracy'); plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.title('Loss'); plt.legend()
plt.tight_layout(); plt.show()
```

```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# === CONFIGURATION ===
X_path = r"X"
aug_path = r"augmented_64"
y_path = r"Y.npy"
feature_names = [
    "Oring", "circular_end_pocket", "v_circular_end_blind_slot",
    "circular_pocket", "non_rectangular_pocket", "rectangular_pocket", "triangular_pocket"
]
VOXEL_SHAPE = (64, 64, 64)

# ---------- helpers ----------
def check_shape(arr, expect=VOXEL_SHAPE):
    return (arr.ndim == 4) and (arr.shape[1:] == expect)

def volume_to_2d_rgb(vol):
    """
    vol: (64,64,64) uint8/float array
    returns (64,64,3) float32 in [0,1]
    Channels: mean projection along Z, Y, X.
    """
    vol = vol.astype(np.float32)
    ch0 = vol.mean(axis=2)          # mean over Z -> (64,64)
    ch1 = vol.mean(axis=1)          # mean over Y -> (64,64)
    ch2 = vol.mean(axis=0)          # mean over X -> (64,64)
    img = np.stack([ch0, ch1, ch2], axis=-1)
    # normalize to [0,1] assuming original was 0..255 (or rescale defensively)
    if img.max() > 1.0:
        img = img / 255.0
    return img.astype(np.float32)

def volumes_to_images(X5d):
    """
    X5d: (N,64,64,64,1) or (N,64,64,64)
    returns (N,64,64,3) float32
    """
    if X5d.ndim == 5:
        X5d = X5d[..., 0]
    imgs = np.empty((len(X5d), 64, 64, 3), dtype=np.float32)
    for i in range(len(X5d)):
        imgs[i] = volume_to_2d_rgb(X5d[i])
    return imgs

# === Load labels ===
y = np.load(y_path)
le = LabelEncoder()
y_encoded = le.fit_transform(y)
num_classes = len(le.classes_)

# === Load perfect data ===
X_list = []
for i, name in enumerate(feature_names):
    f = os.path.join(X_path, f"{i}_{name}_concatenated.npy")
    d = np.load(f)
    if not check_shape(d, VOXEL_SHAPE):
        raise ValueError(f"Shape mismatch in {f}: got {d.shape}, expected (*,64,64,64)")
    X_list.append(d)
X = np.concatenate(X_list, axis=0)[..., np.newaxis]  # (N,64,64,64,1)

# align to labels length
X = X[:len(y_encoded)]

# === Load augmented imperfect data ===
augmented_list = []
aug_labels = []
for i, name in enumerate(feature_names):
    f = os.path.join(aug_path, f"{i}_{name}_voxel_grid_augmented.npy")
    if os.path.exists(f):
        d = np.load(f)
        if not check_shape(d, VOXEL_SHAPE):
            raise ValueError(f"Aug shape mismatch in {f}: got {d.shape}, expected (*,64,64,64)")
        augmented_list.append(d)
        aug_labels.extend([i] * len(d))

if len(augmented_list) > 0:
    augmented_data = np.concatenate(augmented_list, axis=0)[..., np.newaxis]  # (M,64,64,64,1)
    aug_labels = np.array(aug_labels, dtype=np.int64)
else:
    augmented_data = np.empty((0,64,64,64,1), dtype=X.dtype)
    aug_labels = np.empty((0,), dtype=np.int64)

# === Combine perfect + imperfect ===
X_total = np.concatenate([X, augmented_data], axis=0)
y_total = np.concatenate([y_encoded, aug_labels], axis=0)

# safety: length match
if len(X_total) != len(y_total):
    m = min(len(X_total), len(y_total))
    X_total = X_total[:m]
    y_total = y_total[:m]

print("Volumes:", X_total.shape, "Labels:", y_total.shape)

# === Convert volumes -> 2D images (64x64x3) ===
X_img = volumes_to_images(X_total)  # (N,64,64,3), float32 in [0,1]
print("2D images:", X_img.shape)

# === Split ===
X_train, X_test, y_train, y_test = train_test_split(
    X_img, y_total, test_size=0.2, stratify=y_total, random_state=42
)

# === Mixed precision (optional) ===
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy('mixed_float16')  # remove if CPU-only

# === 2D CNN model ===
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(64, 64, 3)),
    tf.keras.layers.Conv2D(32, 3, padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(2),
    tf.keras.layers.Conv2D(64, 3, padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(2),
    tf.keras.layers.Conv2D(128, 3, padding='same', activation='relu'),
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.35),
    tf.keras.layers.Dense(num_classes, activation='softmax', dtype='float32')  # force float32 output for stability
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# === Train ===
history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=64,          # 2D allows bigger batch
    validation_data=(X_test, y_test)
)

# === Plot ===
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('Loss')
plt.legend()
plt.tight_layout()
plt.show()
```

### Augmentation Logic
1. Load .npy file  shape (N, 64, 64, 64)

2. For each voxel grid:

3. Apply shifts along X and Y:

    Left/right (X1, X+1), up/down (Y1, Y+1)

4. Or randomly within a range (e.g., 3)

5. Use np.roll(...) to shift

6. Keep Z untouched.

7. Repeat until we get 3,000 samples per category.

8. Save as a single .npy file per feature.

```python
import os
import numpy as np
from glob import glob

def augment_voxel_grid(grid, num_augments=10, shift_range=3):
    """Generate augmented grids by shifting 0s in x and y (not z)."""
    augmented = []

    for _ in range(num_augments):
        x_shift = np.random.randint(-shift_range, shift_range + 1)
        y_shift = np.random.randint(-shift_range, shift_range + 1)
        
        # Shift only X and Y
        shifted = np.roll(grid, shift=x_shift, axis=0)
        shifted = np.roll(shifted, shift=y_shift, axis=1)

        augmented.append(shifted)

    return augmented

def process_voxel_files(input_dir, output_dir, target_count=3000, shift_range=3):
    os.makedirs(output_dir, exist_ok=True)

    npy_files = glob(os.path.join(input_dir, '*_voxel_grid.npy'))

    for file in npy_files:
        print(f"\n Processing: {os.path.basename(file)}")
        data = np.load(file)  # shape (N, 64, 64, 64)
        print(f" Loaded shape: {data.shape}")

        augmented_all = []
        per_grid_needed = max(target_count // len(data), 1)

        for grid in data:
            augmented = augment_voxel_grid(grid, num_augments=per_grid_needed, shift_range=shift_range)
            augmented_all.extend(augmented)

            if len(augmented_all) >= target_count:
                break  # Stop if enough

        # Truncate if over
        augmented_all = np.array(augmented_all[:target_count], dtype=np.uint8)

        print(f" Generated: {augmented_all.shape}")

        # Save file
        base_name = os.path.splitext(os.path.basename(file))[0]
        out_path = os.path.join(output_dir, base_name + "_augmented.npy")
        np.save(out_path, augmented_all)
        print(f" Saved to: {out_path}")

# === Run this ===
input_folder = r"D:\OneDrive - University of Bristol\Dissertation\Imperfect Data\2_combined_npy"
output_folder = r"D:\OneDrive - University of Bristol\Dissertation\Imperfect Data\3_augmented_shifted"

process_voxel_files(input_folder, output_folder, target_count=3000, shift_range=3)
```
